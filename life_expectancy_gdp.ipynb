```python
# Libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
ax = plt.subplot()
sns.countplot(data=life_expectancy_df, x="Year",ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```

```python
# Bivariate
```

```python
ax = plt.subplot()
sns.boxplot(data=life_expectancy_df, x="Country", y="Life expectancy at birth (years)", ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

Zimbabwe has the worst life expectancy. Between 45 and 55 years, compared to the others that has more than 70 years life expectancy.

```python
ax = plt.subplot()
sns.boxplot(data=life_expectancy_df, x="Year", y="Life expectancy at birth (years)", ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

Life expectancy increased over the years.

```python
ax = plt.subplot()
sns.boxplot(data=life_expectancy_df, x="Country", y="GDP", ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

- United States of America has the greatest GDP.
- Zimbabwe seems to have the lowest GDP.
- China has the most spread GDP.

```python
ax = plt.subplot()
sns.boxplot(data=life_expectancy_df, x="Year", y="GDP", ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

GDP for all the countries in the dataset has become more spread by year.

```python
sns.lmplot(data=life_expectancy_df, x = "GDP", y="Life expectancy at birth (years)")
plt.show()
```

There is a small positive correlation between GDP and life expectancy.

## Conclusions
- USA has the greatest GDP but not the greatest life expectancy.
- Germany has the greatest life expectancy, it is interesant.
- Zimbabwe has the worst GDP and life expectancy.
- There is a small positive correlation between GDP and life expectancy, telling us that GDP increases the life expectancies may increase too.
